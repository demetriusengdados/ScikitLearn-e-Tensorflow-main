```python
#libs
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')
```

```python
### ### ### ### ### ###
# Gerando dados aleatorios
### ### ### ### ### ###
```

```python
n_samples = 30 #numero de amostras(pontos de dados)
n_features = 5 #numero de caracteristicas(dimensões) dos pontos de dados 
n_clusters = 3 #numero de clusters desejados
cluster_std = 5 #Desvio padrão dos clusters (controla a dispersão)
random_state = 10 #semente aleatoria para reprodutibilidade
```

```python
#A função make_blobs gera dados sinteticos com clusters definidos pelos parametros acima 
dados, nada = make_blobs(
    n_samples=n_samples, n_features=n_features, centers=n_clusters,
    cluster_std=cluster_std, random_state=random_state
)
```

```python
#Treinamento do modelo do k-means
modelo_Kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
modelo_Kmeans_labels = modelo_Kmeans.fit_predict(dados)
```

```python
#Treinamento de MOdelo Agglomerative Clustering
Agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
Agg_labels = Agg_clustering.fit_predict(dados)
```

```python
#Gráfico
plt.figure(figsize=(20, 10))
plt.style.use('seaborn-dark')
```

```python
#K-means
plt.subplot(1,3,1)
plt.scatter(dados[:,0], dados[:, 1], c=modelo_Kmeans_labels, cmap='Dark2')
plt.scatter(modelo_Kmeans.cluster_centers_[:,0], modelo_Kmeans.cluster_centers_[:,1], marker='X', s=100, color='red')
plt.title('K-Means')
plt.tight_layout();
```

```python
#Aglometive Clustering
plt.subplot(1,3,2)
plt.scatter(dados[:,0], dados[:,1], c=Agg_labels, cmap='Dark2')
plt.title("Aglomerative")
plt.tight_layout();
```
